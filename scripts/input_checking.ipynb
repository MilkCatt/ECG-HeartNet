```python
import os
import pandas as pd
from tqdm import tqdm

ecg_path = "../data/ecg_clipped"  # Adjust if your path is different

nan_files = []

for fname in tqdm(os.listdir(ecg_path)):
    if not fname.endswith(".csv"):
        continue
    fpath = os.path.join(ecg_path, fname)
    try:
        df = pd.read_csv(fpath)
        if df.isnull().values.any():
            nan_files.append(fname)
    except Exception as e:
        print(f"âš ï¸ Error reading {fname}: {e}")

# Final output
print(f"\nðŸ§¼ Found {len(nan_files)} files with NaNs:\n")
for f in nan_files:
    print(f)

print(nan_files)
```
