```python
import os
import shutil
```

```python
def rename_and_copy_files(source_dir, dest_dir):
    # Create destination directory if it doesn't exist
    os.makedirs(dest_dir, exist_ok=True)
    
    # Walk through the directory structure
    for root, dirs, files in os.walk(source_dir):
        for file in files:
            # Check if file is a CSV and has "JS" in the name
            if file.endswith('.csv') and 'JS' in file:
                # Create new filename without "JS"
                new_filename = file.replace('JS', '')
                
                # Get full paths
                source_file = os.path.join(root, file)
                dest_file = os.path.join(dest_dir, new_filename)
                
                # Copy file to new location with new name
                shutil.copy2(source_file, dest_file)
                print(f"Copied {file} to {new_filename}")
```

```python
source_directory = "data/WFDBRecords"
destination_directory = "processed_data"
    
rename_and_copy_files(source_directory, destination_directory) 
```

```python
import pandas as pd

# Load diagnoses file
diag_path = '../data/diagnoses.csv'
df = pd.read_csv(diag_path)

# Extract numeric part of JS IDs
df['ID_num'] = df['ID'].astype(str).str.extract(r'(\d+)')

# Drop matching rows
to_remove_ids = [f"{int(fname.split('.')[0])}" for fname in files_to_remove]
filtered_df = df[~df['ID_num'].isin(to_remove_ids)]

# Drop the helper column
filtered_df.drop(columns='ID_num', inplace=True)

# Save cleaned version
filtered_df.to_csv('../data/diagnoses_cleaned.csv', index=False)
print(f" Cleaned diagnoses saved  {len(df) - len(filtered_df)} rows removed")
```
